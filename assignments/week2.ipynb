## lambda复习（匿名函数）

```python
def two_number_add(a,b):
    return a+b
two_number_add(1,2)
```

```python
add = lambda a,b:a+b
add(1,2)
```

```python
# 有多少人在他们的工作头衔当中包含Chief这个词？
# .upper()是把所有字母都变成大写
def chief_s(title):
    if "chief" in title.lower():
        return True
    else:
        return False
    
```

```python
# apply 把 数据中的所有值一个一个的给到参数中的函数
df["员工职位"].apply(lambda x : chief_s(x))
```

```python
# 包含 chief 的一共有627
sum(df["员工职位"].apply(lambda x : chief_s(x)))
```
