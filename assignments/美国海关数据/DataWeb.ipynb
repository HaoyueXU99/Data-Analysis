```python
import pandas as pd

# 从Excel表格中读取sheet
df1 = pd.read_excel("dataweb-queryExport_2018_to_2021.xlsx",sheet_name = "2018, General Customs Value",header = 2)
df2 = pd.read_excel("dataweb-queryExport_2018_to_2021.xlsx",sheet_name = "2019, General Customs Value",header = 2)
df3 = pd.read_excel("dataweb-queryExport_2018_to_2021.xlsx",sheet_name = "2020, General Customs Value",header = 2)
df4 = pd.read_excel("dataweb-queryExport_2018_to_2021.xlsx",sheet_name = "2021, General Customs Value",header = 2)

# 将“Data Type”和“HTS Number”列丢弃
df1.drop(["Data Type","HTS Number"],axis = 1,inplace = True)
df2.drop(["Data Type","HTS Number"],axis = 1,inplace = True)
df3.drop(["Data Type","HTS Number"],axis = 1,inplace = True)
df4.drop(["Data Type","HTS Number"],axis = 1,inplace = True)

# 将每个国家中的数据合并（sum），储存
df1 = df1.groupby("Country").sum()
df2 = df2.groupby("Country").sum()
df3 = df3.groupby("Country").sum()
df4 = df4.groupby("Country").sum()

# 初始化新列表
df5=pandas.DataFrame()

# 将列表拼接
df5 = pd.concat([df1,df2,df3,df4], axis=1, ignore_index=False,join='outer')

# 将同一行的数据相加，放进新列
df5["Total"] = df5.iloc[:,0:].sum(axis = 1)

# 根据“Total” 排序
df5.sort_values(by = "Total", ascending = False, inplace = True)

writer = pd.ExcelWriter('dataweb_result.xlsx')                     
df5.to_excel(writer,sheet_name = "汇总")                   

# 保存 Excel
writer.save()
# writer.close()
```
