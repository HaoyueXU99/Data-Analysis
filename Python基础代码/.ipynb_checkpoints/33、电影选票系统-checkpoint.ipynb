```python
class Cinema():
    # 设置用户想看的电影属性
    @property
    def show(self):
        return self.__show
    
    # 设置电影列表
    @property
    def list_film(self):
        return ['战狼','你好，李焕英','人潮汹涌','红海行动']
    # 通过属性设置用户想看的电影
    @show.setter
    def show(self,value):
        self.__show = value
        # 检查用户输入的值是否存在在电影列表中
        if value in self.list_film:
            print("您选择的电影是《{}》".format(value))
        else:
            print("电影名称输入错误，系统退出后，请重新选择")
            exit(0)
    # 设置场次时间属性
    @property
    def video_time(self):
        return self.__video_time
    @video_time.setter
    def video_time(self,value):
        self.__video_time = value
        # 判断时间是否在电影时间列表中
        if value not in self.get_film_time():
            print("场次输入错误，系统退出后请重新输入")
            exit(0)
            
    # 声明一个变量，用来存储用户订票的数量
    @property
    def user_ticket(self):
        print(self.__get_film_time())
    
    # 设置用户订票的属性，定的哪些票
    @user_ticket.setter
    def user_ticket(self,value):
        seat_list = self.get_seat()

        # 这里用户输入的数据进行解析
        user_ticket_list = value.split(",")
        
        # 设置一个状态代表订票是否成功，默认不成功
        status = False
        
        for i in user_ticket_list:
            if i not in seat_list:
                print("输入的座位号不存在，系统退出后请重新选择")
                exit(0)
            # 判断当i等于user_ticket_list中的最后一个元素的时候
            # 代表着循环结束了，并且没有退出程序
            # 设置订票成功状态
            if i == user_ticket_list[-1]:
                status = True
        # 订票成功，打印订票的信息
        if status == True:
            print("恭喜您订票成功，您观看的电影是《{}》,您定的票的作为是{}，您定的场次是{}".format(self.__show,user_ticket_list,self.__video_time))
            
            
    def get_film_time(self):
        return ['8:00',"9:00","10:00"]

    def get_seat(self):
        return ['01',"02","03","04","05","06","07"]
cinema = Cinema()
# 欢迎语
print("欢迎来到大周影院，免费怎公司能够啤酒、饮料、花生米了，外加火腿肠~，再送点爆米花和可乐！")

# 展示电影院中正在上映的电影
print("本影院正在上映的电影有:",cinema.list_film)
# 请用户输入想看的电影，并将其设置成属性
cinema.show = input("请输入您想观看的电影:")

print("电影场次的安排是{}".format(cinema.get_film_time()))
# 请用户选择场次
cinema.video_time = input("请输入观看的场次:")
print("电影的剩余座位是：{}".format(cinema.get_seat()))
# 请用户输入要定的座位号码
cinema.user_ticket = input("请选择您要的座位,多个作为请用逗号(,)进行分隔：")
```
