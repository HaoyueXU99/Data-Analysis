```python
## 嵌套函数：函数内部定义一个函数
def func1():
    print("我是函数1")
    
    def func2():
        print("我是函数2")
    
    func2()
    print("调用函数2完成")

func1()   
```

```python
def func1():
    print("我是函数1")
    
    def func2():
        print("我是函数2")
    
    func2()
    print("调用函数2完成")

    return func2  # 这个是返回函数2的引用
f = func1()  
```

```python
f()
```

```python
# 在一个函数，如func1中的内部定义了另外一个函数func2，并函数1(func1)的返回值是
# 函数func2的引用，这种情况，我们称之为闭包
# 简单说就是外包函数返回内部函数的引用就叫做闭包
```

```python
## 外汇换算程序代码

# 定义一个函数，来实现
def exchange_rate(type,RMB=1):
    # 声明包含的汇率
    compare_func_name = [
        'HKD',"MDP","TWD","EUR","USD","GBP","AUD","KRW","JPY"
    ]
    
    if (type not in compare_func_name):
        print("没有此币种")
        exit(0)
    def HKD():
        return RMB*1.1145
    def MDP():
        return RMB*1.2023
    def TWD():
        return RMB*4.5356
    def EUR():
        return RMB*0.1312
    def USD():
        return RMB*0.1423
    def GBP():
        return RMB*0.1171
    def AUD():
        return RMB*0.2115
    def KRW():
        return RMB*112.7712
    def JPY():
        return RMB*15.0836
    return eval(type)()
print(exchange_rate(type='JPY',RMB=5))
```
