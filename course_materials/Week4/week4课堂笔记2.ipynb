```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
user = pd.read_csv("用户基本信息.csv")
font = {"family":"Arial Unicode MS","size":"20"}
plt.rc("font",**font)

user.head()
```

```python
user.shape
```

```python
chunksize = 1000


user_info_chuck = pd.read_csv("用户基本信息.csv",chunksize=chunksize)
user_info = pd.DataFrame()
start_index = 0
for u in user_info_chuck:
    print("正在处理第" + str(start_index) + "行数据")
    user_info = user_info.append(u.loc[start_index],ignore_index = True)
    start_index += chunksize
print('数据处理完成')
```

```python
user['shopping_level'].unique()
```

```python
user_info.groupby('shopping_level').count()
```

```python
plt.pie(user_info.groupby('shopping_level').count()['age_level'],
       labels = [1,2,3],
       autopct = "%.2f")
plt.title('购物等级占比')
plt.show()
```

```python
# 柱状图绘制各购买等级用户的数量
labels = [1,2,3]
y = user_info.groupby('shopping_level').count()['age_level']

plt.bar(labels,height = y)
# 在每个柱子上显示具体的数量
for a,b in zip(labels,y):
    plt.text(a,b,'%.0f'%b, fontsize = 15)
    
ax = plt.gca()
ax.spines['right'].set_color('white')
ax.spines['top'].set_color('white')

plt.xlabel('购买等级')
plt.ylabel('用户数量')

plt.show()
```

```python
labels = user_info['age_level'].unique()
x = user_info.groupby('age_level').count()['shopping_level']

# 不转为 int 的话 labels 是分阶距离
plt.pie(x,labels = labels.astype(int),
       autopct = "%.0f%%")

plt.title('年龄层次占比')
plt.show()
```

```python
# 创建一个三行六列
x = np.empty((3,0))
x
```

```python
# 按照上边每一列数据生成的逻辑把列表数据赋值
age_level_group_data = user_info.groupby('age_level').count()

for i in age_level_group_data.index.values:
    print("age_level为{}的shopping_level的情况是".format(i))
    result = user_info[user_info['age_level'] == i].groupby('shopping_level').count()['age_level']
    print(result)
    if len(result) != 3:
        print("长度不是3，判断具体哪个值缺失，然后补0")
        arr = [1,2,3]
        
        for i in result.index.values.astype(int):
            print("*"*10)
            print(result[i])
            
            print("-"*10)
            print(i)
            arr.remove(i)
        
        for a in arr:
            print("开始赋值========="+str(a))
            result[float(a)] = 0
            print(result)
        print(arr)
    print("赋值后")
        
    result = result.sort_index()
    print(result)
    x = np.column_stack((x,result))
print(x)
```

```python
plt.imshow(x)

plt.xlabel('年龄层次')
plt.ylabel('购物等级')
plt.xticks(np.arange(6),np.arange(1,user_info['age_level'].unique().astype(int).max()+1))
plt.yticks(np.arange(3),np.arange(1,user_info['shopping_level'].unique().astype(int).max()+1))
plt.colorbar()
plt.show()
```
