# 时间戳是什么
- 时间戳是从1970年1月1日早上8点开始，到现在某一时刻的毫秒值
### 未来要学习mysql，在mysql 中，有时候时间的存储是用时间戳的形式来进行存储
- 时间戳的形式来进行存储所占用的硬盘空间会比时间格式所占用的硬盘空间小
- 为了帮助企业节省那部分硬件资源


```python
dtime = df["订单开始时间"]
dtime
```

```python
## 把时间转换成时间戳
v = (dtime.values-np.datetime64('1970-01-01 08:00:00'))/np.timedelta64(1,'ms')
v
```

```python
df['订单开始时间'] = v
df.head()
```

```python
# 把时间戳再转换为时间
df["订单开始时间"] = pd.to_datetime(df['订单开始时间'],unit='ms',origin = pd.Timestamp('1970-01-01 08:00:00'))
```
