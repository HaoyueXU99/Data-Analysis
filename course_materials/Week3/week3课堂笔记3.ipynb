## 异常值处理
- 主要我们处理的字段是“价格”和“节省”两个字短


```python
df.describe()
```

```python
df.describe().T
```

```python
## 计算价格的标准差
df['价格'].std()
```

```python
df["价格"]-df["价格"].mean()
```

```python
sta = (df["价格"]-df["价格"].mean())/df["价格"].std()
```

```python
sta
```

```python
## 标准分数大于三的我们认为是异常值
df2 = df[sta.abs()>3]
df2
```

```python
### 请把节省的钱和价格的钱的数据筛选出来
df3 = df[df["节省"]>df["价格"]]
```

```python
## 数据拼接
pd.concat([df2,df3])
```

```python
# 从原始数据删除所有的异常值
# 找到异常值的索引
del_index = pd.concat([df2,df3]).index
del_index
```

```python
df.drop(del_index,inplace =True)
```
