```python
## 作业要求
# 龙生九子
# 有一天，九位龙的儿子出游，路遇恶霸一名，实力强大，欺凌弱小，九兄弟决定拔刀相助
# 但每一位都不是恶霸的对手，只能采用车轮战，群起而攻之

# 九兄弟初始血量都是10万，恶霸血量100万，九兄弟每人出手一次后，恶霸出手一次
# 九兄弟，每出手一次，随机造成伤害1到15000
# 如果九兄弟其中有被打到没血了，则退出战场，不能参与下边的战斗了
# 恶霸战死，九兄弟还有剩余，则九兄弟胜，恶霸未死，则恶霸胜
# 恶霸拥有3个技能
# 1、单体攻击，出手一次，随机造成伤害 5000到 10000
# 2、群体攻击，出手一次，给对面所有目标造成伤害，每个目标造成随机伤害1到5000
# 3、恢复技能，当恶霸血量低于10万时，可耗费一次出手机会，使用恢复技能，每次随机恢复10到20万血量

# 当九兄弟只剩下一人时，百分之20的概率召唤出他们的父亲：龙。 发动技能 龙之愤怒，出手一次。造成固定伤害50万。

## 注意：这里我们是恶霸

```

```python
import random
# 龙类
class Dragon:
    # 龙的初始血量
    blood_value = 100000
    # 龙不实现也可以，可以由子类实现
    def fighting(self):
        hurt_value = random.randint(1,15000)
        return hurt_value
    # 龙之愤怒
    @staticmethod
    def dragon_indignation():
        print("何方小儿，伤我子嗣，neng（读四声，这样有气势，等同于'弄'）死你")
        return 500000
    
class QiuNiu(Dragon):
    def __init__(self,name):
        self.name = name
        
    def fighting(self):
        hurt_value = super().fighting()
        print("囚牛发动攻击，老牛的音乐，造成伤害{}".format(hurt_value))
        return hurt_value
class BiXi(Dragon):
    def __init__(self,name):
        self.name = name
        
    def fighting(self):
        hurt_value = super().fighting()
        print("赑屃发动攻击，赑屃之力，造成伤害{}".format(hurt_value))
        return hurt_value
    
class FuXi(Dragon):
    def __init__(self,name):
        self.name = name
        
    def fighting(self):
        hurt_value = super().fighting()
        print("负屃发动攻击，负屃的石碑，造成伤害{}".format(hurt_value))
        return hurt_value
    
class BiAn(Dragon):
    def __init__(self,name):
        self.name = name
        
    def fighting(self):
        hurt_value = super().fighting()
        print("狴犴发动攻击，狴犴虎之力，造成伤害{}".format(hurt_value))
        return hurt_value
    
class YaZi(Dragon):
    def __init__(self,name):
        self.name = name
        
    def fighting(self):
        hurt_value = super().fighting()
        print("睚眦发动攻击，睚眦必报之睚眦的复仇，造成伤害{}".format(hurt_value))
        return hurt_value
    
class ChaoFeng(Dragon):
    def __init__(self,name):
        self.name = name
        
    def fighting(self):
        hurt_value = super().fighting()
        print("嘲风发动攻击，仙人指路，造成伤害{}".format(hurt_value))
        return hurt_value
    
class PuLao(Dragon):
    def __init__(self,name):
        self.name = name
        
    def fighting(self):
        hurt_value = super().fighting()
        print("蒲牢发动攻击，蒲牢巨吼，造成伤害{}".format(hurt_value))
        return hurt_value
    
class SuanNi(Dragon):
    def __init__(self,name):
        self.name = name
        
    def fighting(self):
        hurt_value = super().fighting()
        print("狻猊发动攻击，狻猊的烟火，造成伤害{}".format(hurt_value))
        return hurt_value
    
class ChiWen(Dragon):
    def __init__(self,name):
        self.name = name
        
    def fighting(self):
        hurt_value = super().fighting()
        print("魑吻发动攻击，魑吻的吞噬，造成伤害{}".format(hurt_value))
        return hurt_value  
# 恶霸类
class Villain:
    blood_value = 1000000
    
    def __init__(self,name):
        self.name = name
    # 血量回复技能
    def add_blood_value(self):
        value = random.randint(100000,200000)
        print("{}血量恢复成功，本次恢复{}血量".format(self.name,value))
        self.blood_value += value
        print("血量恢复成功，恢复成功后的血量是{}".format(self.blood_value))
    
    # 单体攻击
    def one_attack(self):
        hurt_value = random.randint(5000,10000)
        return hurt_value
    #群体攻击技能
    def group_attack(self):
        hurt_value = random.randint(1,8000)
        return hurt_value
    
    
```

```python
## 实例化九兄弟
qiuniu = QiuNiu("囚牛")
yazi = YaZi("睚眦")
chaofeng = ChaoFeng("嘲风")
pulao = PuLao("蒲牢")
suanni = SuanNi("狻猊")
bixi = BiXi("赑屃")
bian = BiAn("狴犴")
fuxi = FuXi("负屃")
chiwen = ChiWen("魑吻")

# 把九兄弟放入一个数组，下边攻击的时候方便进行遍历操作
brother_nine_list = [qiuniu,yazi,chaofeng,pulao,suanni,bixi,bian,fuxi,chiwen]

# 实例化恶霸
villain = Villain("帅气的绅士")

# 多态性的应用，创建兄弟攻击的方法
def brother_fighting(obj):
    hurt_value = obj.fighting()
    villain.blood_value -= hurt_value
    return hurt_value

# 声明一个变量，标记谁开始攻击了
who_attack = "恶霸"
```

```python
while True:
    if who_attack == "恶霸":
        # 当恶霸血量低于10万时，询问是否加血
        if villain.blood_value < 100000:
            add_blood_status = input("血量低于10万，是否加血(Y/N)??")
            if add_blood_status.upper() == "Y":
                villain.add_blood_value()
                # 将下一次攻击的机会交给九兄弟
                who_attack = "九兄弟"
                
                continue
        # 循环恶霸使用哪种攻击
        attack_type = input("请问{}使用哪种攻击方式，1为单体攻击，2为群体攻击，请选择".format(villain.name))
        if attack_type == "1":
            # 恶霸使用单体攻击伤害
            villain_one_attack_hurt_value = villain.one_attack()
            # 随机对九兄弟其中之一产生伤害
            one_of_nine = brother_nine_list[random.randint(0,len(brother_nine_list)-1)]
            one_of_nine.blood_value -= villain_one_attack_hurt_value
            print("{}受到攻击，血量剩余{}".format(one_of_nine.name,one_of_nine.blood_value))
            
            if one_of_nine.blood_value < 1:
                print("{}战败身亡，退出战场".format(one_of_nine.name))
                brother_nine_list.remove(one_of_nine)
                print("九兄弟剩余{}个".format(len(brother_nine_list)))
            
        elif attack_type == "2":
            villain_group_hurt_value = villain.group_attack()
            # 九兄弟都要受到伤害
            for i in brother_nine_list:
                i.blood_value -= villain_group_hurt_value
                print("{}受到群体伤害,血量剩余{}".format(i.name,i.blood_value))
                if i.blood_value < 1:
                    print("{}战败身亡，退出战场".format(i.name))
                    brother_nine_list.remove(i)
                    print("九兄弟剩余{}个".format(len(brother_nine_list)))

        # 判断九兄弟是否只剩下一个
        if len(brother_nine_list) == 1:
            print("九兄弟只剩下一个了，请求龙神发动攻击")
            if random.randint(0,5) == 1:
                hurt_value = Dragon.dragon_indignation()
                villain.blood_value -= hurt_value
                if villain.blood_value < 1:
                    print("龙神发威，恶霸阵亡，九兄弟获胜")
            else:
                print("龙神召唤失败")
        
        if len(brother_nine_list) == 0:
            print("九兄弟全部阵亡,{}获胜".format(villain.name))
            break
        # 将下一次攻击交给九兄弟    
        who_attack = "九兄弟"
        
    elif who_attack == "九兄弟":
        for i in brother_nine_list:
            hurt_value = brother_fighting(i)
            print("{0}向{1}发起攻击，造成伤害{2}，{1}的血量剩余{3}".format(i.name,villain.name,hurt_value,villain.blood_value))
            
            if villain.blood_value < 1:
                print("恶霸阵亡，九兄弟获胜")
                break
        who_attack = "恶霸"
    print("="*20)
```
