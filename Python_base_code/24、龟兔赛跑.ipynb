```python
## 装饰器 、语法糖    在java中叫 注解

# 我们统一交装饰器
```

```python
# 装饰器这里需要一个参数,这个参数就是要装饰的函数的名字
def runtime(func): # 其实就是把my_func1传给了func
    
    def wrapper():
        print(1)
        return func()
    
    return wrapper  # 返回包装函数的名字

@runtime   # 这个时候my_func1就使用runtime进行了装饰
def my_func1():
    print("函数1运行")
    return 2
    
r = my_func1()
print(r)
```

```python
# 如果我们想要让这个装饰器具备获取到函数的运行时间功能，应该怎么办
```

```python
# 导入时间模块
import datetime
# 装饰器这里需要一个参数,这个参数就是要装饰的函数的名字
def runtime(func): # 其实就是把my_func1传给了func
    
    def wrapper():
        print(datetime.datetime.now())
        return func()
    
    return wrapper  # 返回包装函数的名字

@runtime   # 这个时候my_func1就使用runtime进行了装饰
def my_func1():
    print("函数1运行")
    return 1

@runtime
def my_func2():
    print("函数2运行")
    return 2

r = my_func1()
print(r)

r2 = my_func2()
print(r2)
```

```python
# 导入时间模块
import datetime
# 装饰器这里需要一个参数,这个参数就是要装饰的函数的名字
def runtime(func): # 其实就是把my_func1传给了func
    
    def wrapper():
        print(datetime.datetime.now())
        r = func()
        print("函数调用完成")
        return r
    
    return wrapper  # 返回包装函数的名字

@runtime   # 这个时候my_func1就使用runtime进行了装饰
def my_func1():
    print("函数1运行")
    return 1

@runtime
def my_func2():
    print("函数2运行")
    return 2

r = my_func1()
print(r)

r2 = my_func2()
print(r2)
```

## 龟兔赛跑代码

```python
# 导入时间模块
import time
import random

# 定义赛跑的跑道长度
track_length = 100

def runtime(func):
    def wrapper():
        start_time = time.time()
        func()
        end_time = time.time()
        print(func.__name__,"运行时间是",end_time-start_time,"秒")
    return wrapper

# 定义乌龟
@runtime
def tortoise():
    for i in range(1,track_length+1):
        print("乌龟跑了{}米".format(i))
        time.sleep(1)

        
# 定义兔子
@runtime
def rabbit():
    # 兔子每跑5米，就随机休息1到10秒
    for i in range(1,track_length+1):
        if i % 5 == 0:
            time.sleep(random.randint(1,10))
        print("兔子跑了{}米".format(i))
tortoise()  
rabbit()
 
```
