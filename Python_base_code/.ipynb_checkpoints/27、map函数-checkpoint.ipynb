## map函数

```python
## map函数支持两个参数，
# 第一个参数是一个函数，就是要处理的逻辑
# 第二个参数是一个可迭代的对象，比如列表就是
```

```python
# 为英雄计算装备属性
def calc_e_properties(e):
    if "红莲斗篷" == e:
        return "防御+120"
    if "魔女斗篷" == e:
        return "魔法防御+360"
    if "影忍之足" == e:
        return "移动速度+60"
    if "辉月" == e:
        return "无敌1.5秒"
    if "破军" == e:
        return "攻击+180"
hero_e = [
    "红莲斗篷",
    "魔女斗篷",
    "影忍之足",
    "辉月",
    "破军"
]
```

```python
p = map(calc_e_properties,hero_e)
print(p)
print(list(p))
```

```python
a = ["a",'d','f']
p = map(lambda x:x.upper(),a)
print(list(p))
```
