```python
def func1():
    return "自信的生命最美丽!"
    print("含泪播种的人必须能含笑收获！")
    
def func2():
    print("勤与反省，才有不断进步的可能")
    
    # yield返回的是一个生成器对象
    yield "一个人，想要优秀，你必须要敢于挑战"
    print("有志者，事竟成，苦心人，天不负")
r1 = func1()
print(r1)

r2 = func2()
print(r2)
```

```python
def func2():
    print("勤与反省，才有不断进步的可能")
    
    # yield返回的是一个生成器对象
    yield "一个人，想要优秀，你必须要敢于挑战"
    print("有志者，事竟成，苦心人，天不负")

# 使用方法
for i in func2():
    print(i)
    print("---------")

print("+++++++++")
r = func2()
for i in r:
    print(i)
    print("******")
```

```python
def func3():
    print("勤与反省，才有不断进步的可能")    
    # yield返回的是一个生成器对象
    yield "生成器第一次返回的值"
    print("有志者，事竟成，苦心人，天不负")
    yield "生成器第二次返回的值"
    print("你不面对现实，现实就会面对你")
    yield "生成器第三次返回的值"
    print("他正在长出翅膀，而我却试图生根")
    
count = 1
for i in func3():
    print("for循环调用生成器之前第{}次打印".format(count))
    print(i)
    count += 1
    print("for循环调用生成器之前第{}次打印".format(count))
    print("*"*20)
```

```python
## 生产的第几个包子
i = 1
# 生产函数
def product():
    global i # 把i声明成为全局变量，这个时候的i变量就可以进行共享操作了
    while i < 101:
        yield "生产的第{}个包子".format(i)
        i += 1
for j in product():
    print(j)
    print("第{}个包子销售了".format(i))
```
