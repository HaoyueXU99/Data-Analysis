```python
# 函数内部调用函数
def function1():
    print("函数1运行")
    return function2()

    
def function2():
    print("函数2运行")
    return 1
r = function1()
print(r)
```

```python
# 函数的返回值是函数的引用
def function1():
    print("函数1运行")
    return function2

    
def function2():
    print("函数2运行")
    return 1
r = function1()
print(r)
# 到这里才是真正的调用了function2函数
z = r()
print(z)
```

## 函数的返回值除了是一个或多个具体的值还可以是函数或者是函数的引用

```python
def function1(i):
    if i == 50:
        return "a"
    
    print("这是函数的第{}次执行".format(i))
    i += 1
    return function1(i)
    
r = function1(1)
print(r)
```

```python
## 21点游戏规则与流程
# 1、玩家共两个角色：电脑和人类，电脑是庄家
# 2、游戏开始时，先给人类和电脑每个玩家分别发两张牌作为底牌，庄家底牌只漏出一张
# 3、判断双方底牌是否直接为21点，如果其中一方为21点则直接判断胜利，
#        并且在总分上加一分。
#      如果双方都是21点，那就是平局，不得分。
# 4、当初识牌面上，没有直接出现21点，人类玩家根据自己的牌面大小决定是否继续要牌。
#       如果要牌，那就在牌堆中抽一张，再次询问是否要牌。
#       如果人类玩家牌面的总点数超过了21点，那就直接判输
# 5、如果人类玩家停止要牌了，并且没有因为超过21点而被判输的情况下，则电脑要牌
#    电脑要牌我们自己来设计一个规则：
#    5.1 比如电脑一直要牌，直到比人类玩家大才停止要牌
#    5.2 根据牌堆中剩余的牌的数量，计算赢的概率，然后设置一个阈值，超过阈值就要，低于就不要。
#    。。。。。。。。。。
#    各种规则都可以，只要起到练习的作用就行
# 6、电脑要完牌之后，直接判断人类玩家和电脑玩家的牌面大小判断胜负
# 7、完成一轮游戏的时候，可由人类玩家决定是否继续玩下一轮
# 8、牌堆中神域的牌数不够玩一轮游戏的时候，游戏自动结束
# 9、计算规则：2、3、4、5、6、7、8、9、10分别是正常的点数，J、Q、K都是10点
# 10、A比较特殊，首先把A当做1来计算，牌面总分数如果小于21，那么再把A当做11再计算一次
#     如果这个时候仍然小于21点，那么A就当11算，如果这个时候牌面总分大于21了，那么A就当1算。



```

```python
# 程序功能模块设计
# 洗牌：将牌进行随机排列 
# 发牌：要牌的时候，需要从牌堆中随机抽取一张牌
# 计分：根据手中的牌的分数进行计算，其中需要考虑A的特殊性
# 胜负判断：当结束要牌的时候，能通过分数判断胜负
# 是否要牌：根据用户产生交互的功能，用户决定是否要牌
# 继续下一轮还是退出
# 游戏结束的判断：让你决定是否提前结束游戏。如果不提前结束
# 则在牌堆中牌的数量不够玩一轮游戏的时候，自动结束游戏

```

```python
playing_cards = {
    "黑桃A": 1, "黑桃2": 2, "黑桃3": 3, "黑桃4": 4, "黑桃5": 5, "黑桃6": 6, "黑桃7": 7,
    "黑桃8": 8, "黑桃9": 9, "黑桃10": 10, "黑桃J": 10, "黑桃Q": 10, "黑桃K": 10,
    "红桃A": 1, "红桃2": 2, "红桃3": 3, "红桃4": 4, "红桃5": 5, "红桃6": 6, "红桃7": 7,
    "红桃8": 8, "红桃9": 9, "红桃10": 10, "红桃J": 10, "红桃Q": 10, "红桃K": 10,
    "方块A": 1, "方块2": 2, "方块3": 3, "方块4": 4, "方块5": 5, "方块6": 6, "方块7": 7,
    "方块8": 8, "方块9": 9, "方块10": 10, "方块J": 10, "方块Q": 10, "方块K": 10,
    "梅花A": 1, "梅花2": 2, "梅花3": 3, "梅花4": 4, "梅花5": 5, "梅花6": 6, "梅花7": 7,
    "梅花8": 8, "梅花9": 9, "梅花10": 10, "梅花J": 10, "梅花Q": 10, "梅花K": 10
}
```

```python
## 导包
import random  # 这个是系统为我们提供的一个函数功能

# 导三方包
# 导入三方包的时候，需要先安装
# 在命令中安装： pip install numpy
# 在Jupyter Notebook中安装三方包 !pip install numpy
# 我们其实不用安装，因为我们使用的是anaconda环境，都已经提前帮我们装好了

# as的作用是起一个别名，后边我们使用到的时候直接使用np就可以了
import numpy as np 

# 这是第二种导包的方式
# exit函数的作用是退出整个程序
from sys import exit
from random import shuffle
```

```python
poker_name = list(playing_cards.keys())
# 声明一个变量用来判断手中的牌是否有A
four_a={"黑桃A","红桃A","梅花A","方块A"}

# 一共使用几副扑克，多副牌就能多玩而一会
poker_count = 1
# 最终生成的牌堆
poker_list = poker_name * poker_count

# 总分的计分器
total_score = np.array([0,0])

# 记录游戏是第几个回合
game_round = 1
# 洗牌：将牌进行随机排列 
def random_card(poker_name):
    shuffle(poker_name)
    return poker_name
# 发牌：要牌的时候，需要从牌堆中随机抽取一张牌
def get_one_poker(poker_list):
    return poker_list.pop(random.randint(0,len(poker_list)-1))
# 计分：根据手中的牌的分数进行计算，其中需要考虑A的特殊性
def score_count(hand_poker):
    # 计算牌的点数
    poker_score = 0
    # 判断是否有A的标记，默认是没有的
    have_a = False
    
    # 先计算手中牌的分数
    for k in hand_poker:
        poker_score += playing_cards[k]
    
    # 判断手中的牌是否有A，再根据A的规则进行分数的计算
    for i in hand_poker:
        if i in four_a:
            have_a = True
            break
        else:
            continue
    if have_a == True:
        if poker_score + 10 <= 21:
            poker_score = poker_score + 10
    return poker_score
# 胜负判断：当结束要牌的时候，能通过分数判断胜负
def who_win(your_score,pc_score):
    # 结束要牌的时候，计算双方点数，判断输赢
    if your_score > 21 and pc_score > 21:
        print('平局了')
        return np.array([0,0])
    elif your_score > 21 and pc_score <= 21:
        print("对不起，你输了")
        return np.array([0,1])
    elif your_score <=21 and pc_score > 21:
        print("恭喜你，你赢了")
        return np.array([1,0])
    elif your_score <= 21 and pc_score <=21:
        if your_score < pc_score:
            print('对不起，你输了')
            return np.array([0,1])
        elif your_score > pc_score:
            print('恭喜，你赢了')
            return np.array([1,0])
        else:
            print('平局了')
            return np.array([0,0])
# 是否要牌：根据用户产生交互的功能，用户决定是否要牌
def if_get_next_poker():
    # 玩家需要判断是否继续要牌
    if_continue = input("是否继续要下一张牌？(Y/N)>>:")
    if if_continue.upper() == "Y":
        return get_one_poker(poker_list)
    elif if_continue.upper() == "N":
        print("你已经停止叫牌")
        return False
    else:
        print("输入错误，请重新输入(Y/N)>>:")
        return if_get_next_poker()
# 继续下一轮还是退出
# 游戏结束的判断：让你决定是否提前结束游戏。如果不提前结束
# 则在牌堆中牌的数量不够玩一轮游戏的时候，自动结束游戏

def continue_or_quit():
    if_next_round = input("你还想要玩下一局吗？(Y/N)>>:")
    if if_next_round.upper() == "Y":
        # 如果继续玩，判断牌堆中的牌数，如果小于10就不让玩了
        if len(poker_list) < 10:
            print("剩余的扑克牌已经不够了")
            print("游戏结束")
            exit(1)
        else:
            return True
    elif if_next_round.upper() == "N":
        print("游戏结束")
        exit(1)
    else:
        print("输入错误，请再输入一次(Y/N)>>:")
        continue_or_quit()

## 开局的时候，自动给玩家和电脑发两张牌
def start_game_init_two_poker(poker_database):
    return [poker_database.pop(random.randint(0,len(poker_database)-1)),
           poker_database.pop(random.randint(0,len(poker_database)-1))]



## 每一轮游戏的过程
def one_round(poker_list):
    # 声明一个变量，存储我们手里的扑克
    your_hand_poker = []
    # 电脑手里的扑克
    computer_hand_poker = []
    # 一个回合的游戏，自动在牌堆中随机抽取两张牌
    you_init_poker = start_game_init_two_poker(poker_list)
    computer_init_poker = start_game_init_two_poker(poker_list)
    
    print("你所获得的扑克牌是{},{}".format(you_init_poker[0],you_init_poker[1]))
    # 电脑只明一张牌，所以第二张牌打印问号
    print("电脑所获得的扑克牌是{},{}".format(computer_init_poker[0],"?"))
    
    # 把获得的扑克牌放到手中
    your_hand_poker.extend(you_init_poker)
    computer_hand_poker.extend(computer_init_poker)
    
    # 计算初识牌面的分数
    score = np.array([score_count(your_hand_poker),score_count(computer_hand_poker)])
    
    # 首先判断一下初始牌面分数的大小，如果有等于21点的，那么直接调用判断输赢函数，游戏结束
    if score[0] == 21 or score[1]==21:
        print("初始牌面分数达到21点")
        return who_win(score[0],score[1])
    # 否则，判断自己手里的分数是否小于21点，如果小于21点那么询问是否要下一张牌
    else:
        while score[0] <= 21:
            get_new_poker = if_get_next_poker()
            
            # 如果get_new_poker不等于False，那就是要了一张牌，并且赋值给了get_new_poker
            if get_new_poker != False:
                # 把新的扑克放进手里
                your_hand_poker.append(get_new_poker)
                print("你手里的扑克牌是{}".format(get_new_poker))
                # 计算分数
                score[0] = score_count(your_hand_poker)
                # 判断分数大小
                if score[0] > 21:
                    print("你手里的扑克牌已经超过21点")
                    print("电脑手里的扑克牌是:{}".format(computer_hand_poker))
                    return who_win(score[0],score[1])
                else:
                    continue
            # 否则，如果没有要下一张牌，那么轮到电脑要牌
            elif get_new_poker == False:
                # 电脑要牌的规则是：只要分数比玩家低就一直叫，直到比玩家分数高就停止
                while score[1] < score[0]:
                    computer_poker = get_one_poker(poker_list)
                    computer_hand_poker.append(computer_poker)
                    # 电脑要完牌之后，计算手中牌的分数
                    pc_score = score_count(computer_hand_poker)
                    score[1] = pc_score
                print("电脑手里的扑克牌是:{}".format(computer_hand_poker))
                return who_win(score[0],score[1])
            else:
                continue
            
while True:
    input("游戏开始，祝你好运！ 按'回车'游戏正式开始\n")
    print("游戏第{}轮".format(game_round))
    # 打印一下游戏的分隔符
    print("*" * 30)
    # 洗牌
    random_card(poker_list)
    print(poker_list)
    # 一轮游戏开始了
    score = one_round(poker_list)
    
    # 计算总比分
    total_score = np.add(total_score,score)
    print("总比分为-- 玩家:{},电脑:{}".format(total_score[0],total_score[1]))
    
    game_round += 1
    continue_or_quit()
```
