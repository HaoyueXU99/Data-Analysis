## while循环

```python
i = 1
while i < 10:
    print(i)
    # 这里需要有一个控制，这个控制就是来控制i的变化方向
    # i有了变化方向，才能够让这个循环停下来
    i = i + 1
```

```python
## 如果循环的中间，我遇到了某一个条件，当这个条件判断成功的时候，我想停止循环
# 代码继续向下走，我应该怎么办？
# break
```

```python
i = 1
while i < 10:
    print(i)
    if i == 5:
        print("i的值现在是5了，我得跳出去了")
        break
    print("i还没有到5，需要加一个值，继续运行")
    i = i + 1
print("代码执行完毕")
```

```python
# break的作用是跳出循环，也就是跳到循环体的外边
```

```python
## 我们一起来开发一个可以计数到100的计数器

# 声明目标值是100
target_number = 10
# 声明一个变量，用来记录初始的值
start_number = 0

while True:
    print(start_number)
    # 这里要控制循环跳出的条件
    # 当我们的开始的值等于了目标的值就跳出循环
    if start_number == target_number:
        print("计数结束")
        break
    start_number = start_number + 1
print("代码运行结束")

```

```python
## 计时器开发需求
# 生活中每60秒等于1分钟，每60分钟等于1个小时
# 开发一个计时器，要求目标时间是3分钟，也就是说计时时间为3分钟，我们的计时器
# 会自动停止
```

```python

# 目标计时时间
target_time = 3700
# 声明一个变量，用来记录初始的时间
keep_time = 0
while True:
    # 首先来控制一下循环的跳出
    if keep_time > target_time:
        print("计时结束")
        break
    # 计算计时时间
    # 当计时时间大于600秒时，时间超过一个小时，将换算成时分秒的形式然后打印
    elif keep_time >= 3600:
        # keep_time对3600取整除，就是小时
        hour = keep_time // 3600
        # keep_time 对3600取余数，然后对60取整除，就是分钟
        minute = keep_time % 3600 // 60
        # keep_time 对3600取余数，然后对60取余数，然后就是秒了
        second = keep_time % 3600 % 60
        print("当前计时时间为"+ str(hour)+"小时"+str(minute)+"分"+str(scond)+"秒")
    # 当计时时间大于等于60秒，并小于3600的时候，将时间换算成分和秒的形式后打印
    elif 3600> keep_time >=60:
        minute = keep_time % 3600 // 60
        scond = keep_time % 3600 % 60
        print("当前计时时间为"+str(minute)+"分"+str(scond)+"秒")
    else:
        print("当前计时时间为"+str(keep_time)+"秒")

    keep_time = keep_time + 1
```
