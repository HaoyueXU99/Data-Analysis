```python
# 这个是摩斯密码的对照表
morse_code_list = [".-","-...","-.-.","-..",".","..-.",
                   "--.","....","..",".---",
                   "-.-",".-..","--","-.","---",".--.",
                   "--.-",".-.","...","-","..-","...-",
                   ".--","-..-","-.--","--.."]
# 从a开始一直到z的字符串
letter_str = "abcdefghijklmnopqrstuvwxyz"
# 我要发送的消息
word_list = ["i", "love",'you']
```

```python
result = ""
for i in word_list:
    print(i)
    for r in i:
        print(r)
```

```python
letter_str.index("c")
```

```python
result = ""
for i in word_list:
    for r in i:
       # 取出来每一个字母在 letter_str中的位置
        word_index = letter_str.index(r)
        # 根据获取到的位置从morse_code_list中获取到莫斯密码的值
        morse_code = morse_code_list[word_index]
        result = result + morse_code + "|"
print(result)
```

```python
# 破译莫斯密码
s = "..|.-..|---|...-|.|-.--|---|..-|"
# 用|分割字符串
word_list = s.split("|")
word_list
```

```python
word_list = word_list[:-1]
word_list
```

```python
result = ""
for i in word_list:
    word_index = morse_code_list.index(i)
    morse_code = letter_str[word_index]
    result = result + morse_code + "|"
print(result)
```
